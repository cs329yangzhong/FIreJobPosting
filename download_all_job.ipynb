```python
import os
import requests
from bs4 import BeautifulSoup
from concurrent.futures import ThreadPoolExecutor, as_completed

def scrape_chronicle_jobs(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            # Assuming you're extracting certain elements, e.g., job descriptions
            job_listings = soup.find_all('div', class_='mds-tabs__panel__content')
            data1 = soup.find_all('div', class_='mds-grid-row')
            
        
            jobs = []
            if job_listings != []:
                jobs = job_listings
            if data1 != []:
                jobs = data1
        
            
            return jobs
        else:
            return []
    except Exception as e:
        print(f"Error scraping {url}: {str(e)}")
        return []
    
    
def save_job_data(job_id, job_content):
    os.makedirs("all_jobs", exist_ok=True)
    file_path = f"all_jobs/id_{job_id}.html"
    with open(file_path, "w") as file:
        for content in job_content:
            file.write(str(content))

def fetch_and_save_job(job_id):
    url = f"https://jobs.chronicle.com/job/{job_id}"
    if not os.path.exists(f"all_jobs/id_{job_id}.html"):
        
        job_content = scrape_chronicle_jobs(url)
        # print(job_content)
        if len(job_content) != 0:
            fields = ["/jobs/business-and-management", "/jobs/education", '/jobs/social-and-behavioral-sciences']
            job_str = str(job_content[0])
            # print(job_str)
            if job_content != [] and "Faculty Positions" in str(job_content[0]) and any([y in job_str for y in fields]):
                # print(job_id)
                save_job_data(job_id, job_content)


# URL of the jobs page
url = "https://jobs.chronicle.com/job/97589"
jobs = scrape_chronicle_jobs(url)
```
